```python
import ezdxf
import pandas as pd
from ezdxf import zoom

file_name='NV-20A.MD'
df=pd.read_csv(f'{file_name}.CSV', usecols=[0,1,2,3,4], names=["est", "x", "y", "z", "des"], header=None)
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

df['des']=df['des'].str.upper()
df['xy']=df.apply(lambda j: (j['x'],j['y']),axis=1)
df['xyz']=df.apply(lambda j: (j['x'],j['y'],j['z']),axis=1)
df.head()
```

```python
#Select only points
filt = (df['des']=='PP')
dfx= df[filt][['est','xyz']]
dfx
```

```python
pps = dfx.values.tolist()
pps
```

```python
# Argument setup=True for adding standard linetypes and text styles.
doc = ezdxf.new('R12',setup=True)
msp=doc.modelspace()

#Config point mode and size
doc.header['$PDMODE'] = 35
doc.header['$PDSIZE'] = 0.5

#Select only points
filt = (df['des']=='PP')
df_pp= df[filt][['est','xyz']]
pps = df_pp.values.tolist()

#Draw points
for pp in pps:
    msp.add_point(pp[1])
    msp.add_text(pp[0],
                dxfattribs={
                    'style': 'LiberationSerif',
                    'height': 0.5}
                ).set_pos(pp[1], align='LEFT')

zoom.extents(msp)
doc.saveas('points.dxf')
```

```python
import ezdxf
import pandas as pd
from ezdxf import zoom

file_name='NV-20A.MD'
df=pd.read_csv(f'{file_name}.CSV', usecols=[0,1,2,3,4], names=["est", "x", "y", "z", "des"], header=None)
df['xy']=df.apply(lambda j: (j['x'],j['y']),axis=1)
df['xyz']=df.apply(lambda j: (j['x'],j['y'],j['z']),axis=1)

# Argument setup=True for adding standard linetypes and text styles.
doc = ezdxf.new('R12',setup=True)
msp=doc.modelspace()

#Config point mode and size
doc.header['$PDMODE'] = 35
doc.header['$PDSIZE'] = 0.5

#Select only points
filt = (df['des']=='PP')
df_pp= df[filt][['est','xyz']]
pps = df_pp.values.tolist()

#Draw points
for pp in pps:
    msp.add_point(pp[1])
    msp.add_text(pp[0],
                dxfattribs={
                    'style': 'LiberationSerif',
                    'height': 0.5}
                ).set_pos(pp[1], align='LEFT')

zoom.extents(msp)
doc.saveas('points.dxf')
```
