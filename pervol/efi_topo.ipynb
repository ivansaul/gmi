## **Cleaning the data**

```python
df.drop([0,1], inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

- Starting with the two new rows. Format and create the new header for the dataframe
    + Row 0 -> Dates
    + Row 1 -> Work shifts

```python
#Row 0
r0=['']*7
for i in df.iloc[0][7:]:
    if i is np.nan:
        #print(True)
        r0.append(True)
    else:
        r0.append(i.strftime("%d/%m/%Y"))
        #print(i.strftime("%d/%m/%Y"))

for i in range(len(r0)):
    if r0[i] is True:
        r0[i]=str(r0[i-1])
    else:
        r0[i]=str(r0[i])

#Row 1
r1=df.iloc[1].tolist()

#New row (header)
rx=[]
for i,j in zip(r0,r1):
    #print(f'{i} {j}'.strip())
    rx.append(f'{i} {j}'.strip())

#rx
```

```python
df.set_axis(rx, axis=1, inplace=True)
df.drop([0,1], inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

## **Implementation**

```python
df_avances=pd.DataFrame(columns=['NIVEL', 'LABOR', 'AVANCE', 'FECHA'])

for i in range(df.shape[0]):
    dfx=df.iloc[[i]].dropna(axis='columns',how='any')
    
    medida = dfx.values.tolist()[0][7:]
    fecha = dfx.columns.tolist()[7:]
    if len(medida) > 0:
        nivel=[dfx['NIVEL'].iloc[0]]*len(medida)
        labor=[dfx['LABOR'].iloc[0]]*len(medida)
        avances=list(zip(nivel,labor,medida,fecha))
        dfx= pd.DataFrame(data=avances, columns=['NIVEL', 'LABOR', 'AVANCE', 'FECHA'])
        #print(dfx)
        df_avances = df_avances.append(dfx, ignore_index=True)
        #print(df_avances)

df_avances
```

to_csv

```python
def guardia(f):
    g=f.split(' ')[1]
    if g=='D': 
        return 'DÍA'
    else: 
        return 'NOCHE'
```

```python
df_avances['GUARDIA']=df_avances['FECHA'].apply(guardia)
df_avances['FECHA'] = df_avances['FECHA'].apply(lambda x: x.split(' ')[0])
df_avances
```

```python
df_avances.to_excel('efi_topo.xlsx', index=False, encoding='utf-8')
```
