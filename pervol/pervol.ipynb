```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('EFICIENCIA MARZO TOPOGRAFÍA - 2022.xlsx', sheet_name='EFICIENCIA FEBRERO')
df.drop([0,1], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
#Row 0
r0=[]
for i in df.iloc[0][7:]:
    if i is np.nan:
        #print(True)
        r0.append(True)
    else:
        r0.append(i.strftime("%d/%m/%Y"))
        #print(i.strftime("%d/%m/%Y"))

for i in range(len(r0)):
    if r0[i] is True:
        r0[i]=str(r0[i-1])
    else:
        r0[i]=str(r0[i])

r0=['']*7+r0

#Row 1
r1=df.iloc[1].tolist()

#New row
rx=[]
for i,j in zip(r0,r1):
    #print(f'{i} {j}'.strip())
    rx.append(f'{i} {j}'.strip())

#rx
```

```python
df.set_axis(rx, axis=1, inplace=True)
df.drop([0,1], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
# Cleaning data
df.dropna(axis='index',how='any',subset=['NIVEL', 'LABOR'],inplace=True)
df.shape
```

```python
# Change data type
df = df.astype({"NIVEL": str})

# delete unnecesary rowns
ur=[]
for index, row in df.iterrows():
    if len(row['NIVEL'])>4:
        print(row['NIVEL'], index)
        ur.append(index)
df.drop(ur, inplace=True)
df.reset_index(drop=True, inplace=True)
```

## **Implementation**

```python
df_avances=pd.DataFrame(columns=['NIVEL', 'LABOR', 'AVANCE', 'FECHA'])

for i in range(df.shape[0]):
    dfx=df.iloc[[i]].dropna(axis='columns',how='any')
    
    fecha = dfx.values.tolist()[0][7:]
    avance = dfx.columns.tolist()[7:]
    if len(avance) > 0:
        nivel=[dfx['NIVEL'].iloc[0]]*len(avance)
        labor=[dfx['LABOR'].iloc[0]]*len(avance)
        avances=list(zip(nivel,labor,fecha,avance))
        dfx= pd.DataFrame(data=avances, columns=['NIVEL', 'LABOR', 'AVANCE', 'FECHA'])
        #print(dfx)
        df_avances = df_avances.append(dfx, ignore_index=True)
        #print(df_avances)

df_avances
```

```python
df_avances.to_csv('pv.csv', index=False)
```
